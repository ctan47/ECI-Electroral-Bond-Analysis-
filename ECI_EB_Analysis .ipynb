```python
import pdfplumber
import pandas as pd

# Open the PDF file
with pdfplumber.open("PF1.pdf") as pdf:
    data = []
    # Iterate over each page
    for page in pdf.pages:
        # Extract text from the page
        text = page.extract_text()
        # Split text by lines
        lines = text.split('\n')
        # Loop through each line
        for line in lines:
            # Assuming the format of each line is "Date of Purchase - Purchaser Name - Denomination"
            elements = line.split('-')
            if len(elements) == 3:
                data.append({'Date of Purchase': elements[0].strip(),
                             'Purchaser Name': elements[1].strip(),
                             'Denomination': elements[2].strip()})
            
# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
import pdfplumber

# Open the PDF file
with pdfplumber.open("PF1.pdf") as pdf:
    # Initialize a counter for lines read
    lines_read = 0
    # Initialize a dictionary to store denominations for each company
    company_sum = {}
    
    # Iterate over each page
    for page in pdf.pages:
        # Extract text from the page
        text = page.extract_text()
        # Split text by lines
        lines = text.split('\n')
        # Loop through each line
        for line in lines:
            # Increment the counter for lines read
            lines_read += 1
            # Split the line into elements
            elements = line.split('-')
            # Check if the line contains valid data
            if len(elements) == 3:
                purchaser_name = elements[1].strip()
                denomination = float(elements[2].strip()) / 100000  # Divide by 100,000
                # Add denomination to the corresponding purchaser/company
                if purchaser_name in company_sum:
                    company_sum[purchaser_name] += denomination
                else:
                    company_sum[purchaser_name] = denomination
            # Check if 100 lines have been read
            if lines_read >= 100:
                break
        # Check if 100 lines have been read
        if lines_read >= 100:
            break

# Print the sum of denominations for each company
for company, total_denomination in company_sum.items():
    print(f"{company}: {total_denomination}")
```

```python
import pandas as pd

df1 = pd.read_csv("purchaser.csv")


company_sum = df1.groupby('Purchaser Name')['Denomination'].sum()


company_sum_sorted = company_sum.sort_values(ascending=False)

print(company_sum_sorted)
```

```python
import pandas as pd

df2 = pd.read_csv("encasher.csv")


party_sum = df2.groupby('Name of the Political Party')['Denomination'].sum()

party_sum_sorted = party_sum.sort_values(ascending=False)

print(party_sum_sorted)
```

```python
import matplotlib.pyplot as plt

df1['Denomination'] = df1['Denomination'] / 10000000  

top_10_companies = company_sum_sorted.head(10)

plt.figure(figsize=(10, 6))
top_10_companies.plot(kind='bar')
plt.title('Top 10 Companies by Sum of Denominations')
plt.xlabel('Company')
plt.ylabel('Sum of Denominations (in crores)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
